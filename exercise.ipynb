```python
import csv

data = [
    {'name': 'Alice', 'age': 30, 'city': 'Seoul'},
    {'name': 'Bob', 'age': 25, 'city': 'Busan'}
]

with open('data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['name', 'age', 'city']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for row in data:
        writer.writerow(row)
```

```python
import xml.etree.ElementTree as ET

root = ET.Element('people')

person1 = ET.SubElement(root, 'person')
ET.SubElement(person1, 'name').text = 'Alice'
ET.SubElement(person1, 'age').text = '30'
ET.SubElement(person1, 'city').text = 'Seoul'

person2 = ET.SubElement(root, 'person')
ET.SubElement(person2, 'name').text = 'Bob'
ET.SubElement(person2, 'age').text = '25'
ET.SubElement(person2, 'city').text = 'Busan'

tree = ET.ElementTree(root)
tree.write('data.xml', encoding='utf-8', xml_declaration=True)
```

```python
import requests
from bs4 import BeautifulSoup
import pymysql

# 데이터베이스 연결 설정
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='rlaekqls23',
    db='mydatabase',
    charset='utf8mb4'
)

try:
    with connection.cursor() as cursor:
        # 크롤링 대상 URL
        url = 'https://www.example.com/articles'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # 기사 목록 추출
        articles = soup.find_all('div', class_='article')

        for article in articles:
            title = article.find('h2').get_text()
            author = article.find('span', class_='author').get_text()
            content = article.find('div', class_='content').get_text()
            pub_date = article.find('span', class_='date').get_text()

            # 데이터베이스에 저장
            sql = """
            INSERT INTO articles (title, author, content, publication_date)
            VALUES (%s, %s, %s, %s)
            """
            cursor.execute(sql, (title, author, content, pub_date))

    connection.commit()
finally:
    connection.close()
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.hyundai.com/kr/ko/digital-customer-support/helpdesk/faq")

# 스크롤 다운하여 상품 로드
for _ in range(5):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(1.5)

# 상품 리스트 가져오기
products = driver.find_elements(By.CSS_SELECTOR, ".ui_accordion.acc_01 dl")

print("총 상품 수:", len(products))

for product in products:
    try:
        element = driver.find_element(By.CSS_SELECTOR, "button.btn")
        element.click()
        link_line = product.find_element(By.CSS_SELECTOR, ".link_line").text
    except:
        link_line = "업체명 없음"

    try:
        wrap_word = product.find_element(By.CSS_SELECTOR, ".listView > tr > td:nth-child(2)").text
    except:
        wrap_word = "업체구분 없음"

    try:
        tel = product.find_element(By.CSS_SELECTOR, ".flex-item .pc_only").text
    except:
        tel = "가격 없음"

    print(f"상품명: {link_line} | 가격: {wrap_word} | 전화번호: {tel}")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.implicitly_wait(2)
wait = WebDriverWait(driver, 10)

driver.get("https://www.hyundai.com/kr/ko/digital-customer-support/helpdesk/faq")

def safe_click(el):
    """스크롤 후 JS 클릭을 시도 (일반 click 대신 사용)."""
    try:
        driver.execute_script("arguments[0].scrollIntoView({block:'center'});", el)
        time.sleep(0.15)
        driver.execute_script("arguments[0].click();", el)
        time.sleep(0.15)
    except Exception:
        # 마지막 수단: 포인터 이벤트 제거 후 클릭 시도
        try:
            driver.execute_script("""
                arguments[0].style.pointerEvents = 'auto';
                arguments[0].click();
            """, el)
            time.sleep(0.15)
        except Exception:
            pass

def ensure_on(dt_element, timeout=2.0):
    """
    dt_element에 'on' 클래스가 생길 때까지 시도.
    1) 이미 on이면 True 반환
    2) safe_click 시도
    3) 짧게 대기하면서 class 체크
    4) 실패 시 JS로 클래스 강제 추가 및 dd 보이기 시도 (폴백)
    """
    start = time.time()
    if "on" in (dt_element.get_attribute("class") or ""):
        return True

    # 1차 시도: 안전한 클릭
    safe_click(dt_element)

    # 2차: on 클래스 올 때까지 짧게 반복 검사
    while time.time() - start < timeout:
        if "on" in (dt_element.get_attribute("class") or ""):
            return True
        time.sleep(0.1)

    # 3차 폴백: JS로 클래스 강제 추가하고 dd를 열기
    try:
        driver.execute_script("""
            var dt = arguments[0];
            dt.classList.add('on');
            var dd = dt.nextElementSibling;
            if(dd) {
                dd.style.display = 'block';
                // 만약 내부 exp가 숨겨져 있다면 강제로 텍스트 추출 가능하도록
                var exp = dd.querySelector('.exp');
                if(exp) exp.style.display = 'block';
            }
        """, dt_element)
        time.sleep(0.1)
        return "on" in (dt_element.get_attribute("class") or "")
    except Exception:
        return False

def crawl_current_page():
    # 페이지 로딩/렌더 안정화를 잠깐 둠
    time.sleep(0.8)

    # 모든 dl 요소를 가져옴 (페이지 내 순서 유지)
    dls = driver.find_elements(By.CSS_SELECTOR, ".ui_accordion.acc_01 dl")
    print(f"이 페이지 dl 수: {len(dls)}")

    for idx, dl in enumerate(dls, start=1):
        # dl 내 첫 dt (요청하신 'dl>dt' 단위 처리)
        try:
            dt = dl.find_element(By.TAG_NAME, "dt")
        except Exception:
            print(f"[{idx}] dl에 dt가 없음, 건너뜀")
            continue

        # 1~2번: dt에 on 있는지 확인, 없으면 추가(클릭) 후 크롤링
        is_on = ensure_on(dt, timeout=2.0)
        if not is_on:
            print(f"[{idx}] WARNING: dt에 on 클래스 추가 실패했지만 계속 진행합니다.")

        # 제목 추출
        try:
            title_el = dt.find_element(By.CSS_SELECTOR, ".title")
            title = title_el.text.strip()
        except Exception:
            title = "제목 없음"

        # dd(내용) 추출 — dt 바로 다음 형제 dd[1]
        try:
            dd = dt.find_element(By.XPATH, "following-sibling::dd[1]")
            # .exp 안에 내용이 있다면 가져온다
            try:
                exp = dd.find_element(By.CSS_SELECTOR, ".exp").text.strip()
            except:
                # .exp가 없거나 비어있다면 dd 전체 텍스트로 대체
                exp = dd.text.strip()
            if not exp:
                exp = "내용 없음"
        except Exception:
            exp = "내용 없음"

        print("-" * 60)
        print(f"[{idx}] 제목: {title}")
        print(f"[{idx}] 내용: {exp[:300]}")  # 너무 길면 앞부분만 출력
        print("-" * 60)

# 전체 페이지 순회
while True:
    crawl_current_page()
    time.sleep(0.6)

    # 다음 페이지 버튼을 찾아 클릭 (disabled 검사 포함)
    try:
        next_btn = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "button.navi.next")))
    except Exception:
        print("다음 버튼을 찾지 못함 — 종료")
        break

    # disabled이면 종료
    if "disabled" in (next_btn.get_attribute("class") or ""):
        print("마지막 페이지입니다. 종료합니다.")
        break

    # 다음 페이지 클릭 (안정 클릭 사용)
    try:
        safe_click(next_btn)
        # 페이지 전환 애니메이션/로딩 안정화
        time.sleep(1.2)
    except Exception as e:
        print("다음 페이지 클릭 실패:", e)
        break

# 종료
driver.quit()
```
